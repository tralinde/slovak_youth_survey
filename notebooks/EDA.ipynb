## Conducting EDA on the data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
youth_data = pd.read_csv('../data/responses.csv')
youth_data.head()
```

```python
youth_data.describe()
```

```python
all_columns_sum = youth_data.sum()
print(all_column_sum)
```

```python
# Select only integer columns
integer_columns = youth_data.select_dtypes(include='float')

# Calculate the sum
most_popular_responses = integer_columns.sum().sort_values(ascending=False)

print(most_popular_responses.head(30))
```

```python
least_popular_responses = integer_columns.sum().sort_values(ascending=True)

print(least_popular_responses.head(30))
```

Consistent with my original proposal, I could begin to filter out the respondents who list "Writing", "Gardening" and "Charity" as their primary activity.

```python
filter_criteria = (youth_data['Writing'] == 5) | (youth_data['Gardening'] == 5) | (youth_data['Charity'] == 5)

# Apply the boolean mask to filter the dataset
outcast_group = youth_data[filter_criteria]

# Print the filtered dataset
outcast_group.describe()
```

```python
outcast_interests = outcast_group.select_dtypes(include='float').sum().sort_values(ascending=False)

print(outcast_interests.head(30))
```

This return seems to suggest that the niche respondents that I chose to isolate are rather conventional with "music" being the aggregate key interest of this group.  There might be more insights from this based upon personality data but this is likely too subjective from a collection of self-reported surveys.
